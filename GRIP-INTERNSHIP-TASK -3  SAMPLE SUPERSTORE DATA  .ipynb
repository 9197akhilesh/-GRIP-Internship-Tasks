# Data Visualization

```python
plt.figure(figsize  = (15,9))
sns.countplot(x='Ship Mode', data=ss_df)
plt.show()
```

```python
plt.figure(figsize  = (15,9))
sns.countplot(x='Segment', data=ss_df)
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.countplot(x=ss_df['State'])
plt.xticks(rotation=90)
plt.title('Count of State wise shipments')
plt.show()
```

# Visualization Understanding of Count of State Wise Shipments :
State wise California has highest sales.

States like Texas, New York, etc are active in business.

Few states like North Dakota, Wyoming and West Virginia have less or no sales at all.

To generate more profit we have to increase our sales in Texas and New York where we have potential customers.

```python
plt.figure(figsize = (15,9))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(ss_df['Category'].value_counts(), labels=ss_df['Category'].value_counts().
        index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

# This pie graph indicates that the Office supplies have the highest demand and others are comparatively less in demand

```python
plt.figure(figsize = (15,9))
ss_df.groupby('Region')['Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

# Overview Understannding of Profit in different Region Chart

West generates most of the profit followed by East.

Central and Southern region are generating less profit as compared to the Eastern and Western Region.

Increase the sales in Central and Southern region in order to generate more profit in these region.

```python
sp = ss_df.groupby('State')['Profit'].sum()
plt.figure(figsize=(15,9))
sns.barplot(x=sp.index, y=sp.values)
plt.xticks(rotation=90)
plt.ylabel('Profit')
plt.show()
```

# This shows that Furniture is generating very less profit.

```python
plt.figure(figsize=(15,9))
sns.countplot(x='Category',hue='Segment',data=ss_df)
plt.ylabel('Count of Segment')
plt.title('Count of Segment by Category and Segment')
```

# The highest purchases are in consumer segment whereas home office segment has the least purchases in each category. 

# We need to focus on the corporate and home office segment as well to increase our sales and profit.

```python
ss_df.groupby('Sub-Category')[['Sales','Profit']].agg(sum).plot(kind='barh',figsize=(15, 9))
plt.ylabel('Sub-Category')
plt.xlabel('Sales and Profit')
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.countplot(x='Region',hue='Ship Mode',data=ss_df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of ship mode by region and ship mode')
```

# Least sales are in Southern region, we can attract this region by offering more discount. 

# Standard mode of shipment is opted by most of the customers in each region.

```python
ss_df.plot(kind = "scatter",figsize = (15,9), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(ss_df['Discount'], ss_df['Profit'], data=ss_df)
```
